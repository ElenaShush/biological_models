```python
import numpy as np
import scipy as sc
import scipy.integrate as slv
import matplotlib.pyplot as plt
```

## Уравнение взрывного роста

$$ 
\frac{\partial y}{\partial t} = y^2 - y
$$







### 1. Точное решение уравнения.

$$ \frac{\partial y}{\partial t} = y^2 - y $$

Это однородное уравнение с разделяемыми переменными. Делим левую часть на $y^2 - y$ и интегрируем.

$$ \frac{dy}{y^2 - y} = dt $$
$$ \int \frac{dy}{y^2 - y} = \int dt $$

Правая часть интегрируется без проблем. Отдельно интегрируем левую часть.

$$ \int \frac{dy}{y^2 - y} = \int \frac{dy}{(y-1)y} = \int \frac{y - (y-1)}{(y-1)y}dy = \int \left ( \frac{dy}{y-1} - \frac{dy}{y} \right ) = \int \frac{dy}{y-1} - \int \frac{dy}{y} = \int \frac{d(y-1)}{y-1} - \int \frac{dy}{y} = ln(|y-1|) - ln(|y|) + C $$

Подставляем полученное решение в дифференциальное уравнение.

$$ ln(|y-1|) - ln(|y|) = t + C $$
$$ ln(|\frac{y-1}{y}|) = t + C $$
$$e^{ln(|\frac{y-1}{y}|)} = e^{t + C} $$
$$ \frac{y-1}{y} = Ce^t $$
$$ \frac{1}{y} = 1 - Ce^t $$
$$ y = \frac{1}{1 - Ce^t} $$

При С = 0 решение становится тривиальным: $ y = 1 $. 



---



График функции при C > 0, например, C = {1, 2, 3}

$$ y = \frac{1}{1 - e^t} $$






```python
Fa = lambda a, t: 1/(1 - a*np.exp(t))

tt = np.linspace(-10, 10, 1000)

plt.figure(figsize=(7, 5))
plt.xlim(-10, 10)
plt.xlabel('t')
plt.ylabel('a')
plt.ylim(-20, 20)

for C in [1, 2, 3]:
  label = "C = "+str(C)
  plt.plot(tt, Fa(C, tt), label=label)
  #plt.plot(sol.t, sol.y[0], label='RK')

plt.legend()

plt.hlines(0, -20, 10, color="grey", linewidth=1, linestyle="--")
plt.vlines(0, -20, 20, color="grey", linewidth=1, linestyle="--")
None
```

График функции при C < 0, например, С = -1, -2, -3

$$ y = \frac{1}{1 + e^t} $$

```python
tt = np.linspace(-8, 6, 1000)

plt.figure(figsize=(7, 5))
plt.xlim(-8, 6)
plt.xlabel('t')
plt.ylabel('a')
plt.ylim(-0.25, 1.25)

for C in [-1, -2, -3, -4, -10, -20]:
  label = "C = "+str(C)
  plt.plot(tt, Fa(C, tt), label=label)

plt.legend()

plt.hlines(0, -8, 6, color="grey", linewidth=1, linestyle="--")
plt.vlines(0, -0.25, 1.5, color="grey", linewidth=1, linestyle="--")
None
```

### Решение явным методом и не явным способом.

Сначала отрисовываем график явного решения дифференциального уравнения.

```python
tt = np.linspace(-6, 6, 1000)

func_decision = lambda t: t**2 - t

plt.figure(figsize=(7, 5))
plt.xlim(-8, 8)
plt.xlabel('t')
plt.ylabel('dy/dt')
plt.ylim(-5, 20)
plt.plot(tt, func_decision(tt))

plt.hlines(0, -8, 8, color="grey", linewidth=1, linestyle="--")
plt.vlines(0, -5, 20, color="grey", linewidth=1, linestyle="--")
None
```

```python
explosive_growth = lambda _, y: y**2 - y

f = lambda _, N: explosive_growth(N)
y0 = 3
atol = 10
rtol = 10

plt.figure(figsize=(21, 5))
#plt.subplots(1, 3)

plt.subplot(121)
plt.xlim(-2, 8)
plt.xlabel('t')
plt.ylabel('a')
plt.ylim(-25, 30)

for y0 in [1.5, 2, 5]:
  yo_label = ", y0 = "+str(y0)
  sol_explicit = slv.solve_ivp(explosive_growth, [0.01, 10.0], [y0], method='RK45', atol=atol, rtol=rtol, first_step=0.2)
  sol_implicit = slv.solve_ivp(explosive_growth, [0.01, 10.0], [y0], method='BDF', atol=atol, rtol=rtol, first_step=0.2 )
  plt.plot(sol_explicit.t, sol_explicit.y[0], label='RK'+yo_label)
  plt.plot(sol_implicit.t, sol_implicit.y[0], label='BDF'+yo_label)

plt.legend()

plt.subplot(122)
'''
plt.xlim(-2, 8)
plt.xlabel('t')
plt.ylabel('a')
plt.ylim(-25, 30)
'''

t_range = [0.01, 10.0]
for y0 in [0.25, 0.5, 0.75]:
  yo_label = ", y0 = "+str(y0)
  sol_explicit = slv.solve_ivp(explosive_growth, t_range, [y0], method='RK45', atol=atol, rtol=rtol, first_step=0.2)
  sol_implicit = slv.solve_ivp(explosive_growth, t_range, [y0], method='BDF', atol=atol, rtol=rtol, first_step=0.2 )
  plt.plot(sol_explicit.t, sol_explicit.y[0], label='RK'+yo_label)
  plt.plot(sol_implicit.t, sol_implicit.y[0], label='BDF'+yo_label)

plt.legend()
None
```

## Популяционная модель Олли

$$ \frac{\partial N}{\partial t} = rN \left (1 - \frac{N}{K} \right )(N - l) $$

где 0 < *l* < *K*, *r* > 0.


3. Построим сначала график правой части уравнения. Из самой формулы видно, что точками должны быть *N = 0, N = K и N = l*, в этих точках функция обращается в ноль.

```python
olli = lambda N,r,K,l: r*N*(1 - N/K)*(l - N)

r = 1
l = 10
K = 20
tt = np.linspace(-4, 22, 1000)
f = lambda N: olli(N, r, K, l)

plt.figure(figsize=(10, 7))
#plt.xlim(-8, 8)
plt.xlabel('t')
plt.ylabel('N')
plt.ylim(-25, 23)
plt.plot(tt, f(tt))
plt.hlines(0, -5, 22, color="grey", linewidth=1, linestyle="--")

plt.scatter(0, 0, color='red')

plt.scatter(l, 0, color='red')
plt.text(l+0.5, 1, "l", fontsize="xx-large")

plt.scatter(K, 0, color='red')
plt.text(K-1, 1, "K", fontsize="xx-large")


dx_arraw = 0.5
xstep = 1.5
y_arraw = -1
plt.arrow(-dx_arraw, y_arraw, -1.5, 0, width = 0.2)
plt.arrow(dx_arraw, y_arraw, 1.5, 0, width = 0.2)

plt.arrow(l-1.5-xstep, y_arraw, xstep, 0, width = 0.2)
plt.arrow(l+1.5+xstep, y_arraw, -xstep, 0, width = 0.2)

plt.arrow(K-dx_arraw, y_arraw, -1.5, 0, width = 0.2)
plt.arrow(K+dx_arraw, y_arraw, 1.5, 0, width = 0.2)
        

None
```

Из графика видно, что точка N0 = 0 является стационнарной, N = 0. Что, конечно, логично, т.к. если размер популяции изначально равен 0, то и изменений размера популяции не будет. 

Точка N0 = l является стационарной и сходящейся, размер популяции будет сходится к размеру l.

При N0 = K модель становится расходящейся, при N0 > K размер популяции будет неограниченно расти, при l < N0 < K размер популяции будет сходиться к l.

### 2. Построим пучки решений.

Сначала проварьируем начальные условия N0 так, чтобы N0 последовательно входило в три промежутка: (0, l), (l, K) и $(K, ∞)$.


```python
r = 1
l = 10
K = 20

atol = 1
rtol = 1
t_range = [0, 4.0]
f_olli_solv = lambda _, N: olli(N, r, K, l)


plt.figure(figsize=(10, 7))
plt.xlabel('t')
plt.ylabel('N')

for N0 in [5, 15, 23, 30]:
  label = "N0 = "+str(N0)
  olli_sol = slv.solve_ivp(f_olli_solv, t_range, [N0], method='BDF', 
                         atol=atol, rtol=rtol, first_step=0.1)
  plt.plot(olli_sol.t, olli_sol.y[0], label=label)


plt.ylim(-2, 30)
plt.hlines(0, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.legend()
None
```

Вопреки ожиданиям модель при N0 = 23 сошлась к N = 10 (N = l, т.к. l = 10). Видно, что на первом шаге график пошел вверх, как и ожидалось из анализа, затем развернулся и пришел к линии N = 10. Кажется, здесь проблема в методе расчета, он не подходит для этой функции. 

При N0 = 30 размер популяции должен уйти вверх, на графике он почему-то ушел вниз. 

Посмотрим на пучки решений при N0 = 15, при том же l = 10, но проварьируем K. Разумно предположить, что для такого значения N0 это даст разную скорость схождения к стационарному решению. Проверим это на графике.

```python
plt.figure(figsize=(10, 7))
plt.xlabel('t')
plt.ylabel('N')

olli_var = lambda _, N, K: olli(N, K=K)
N0 = 15

for K in [16, 20, 50, 100]:
  olli_var = lambda _,N: olli(N, r, K, l)
  label = "K = "+str(K)
  olli_sol = slv.solve_ivp(olli_var, t_range, [N0], method='BDF', 
                         atol=atol, rtol=rtol, first_step=0.1)
  plt.plot(olli_sol.t, olli_sol.y[0], label=label)


plt.ylim(0, 16)
plt.legend()
None
```

Из графика видно, что чем больше K, точнее, чем больше K отличается от l, тем быстрее сходится решение.

Теперь зафиксируем N0 = 15, K = 20 и проварьируем l. Т.к. N0 находится между l и K, это даст разные варианты сходимости и, видимо, разную скорость схождения.

```python
plt.figure(figsize=(10, 7))
plt.xlabel('t')
plt.ylabel('N')

N0 = 15
K = 20

for l in [2, 5, 10, 14]:
  olli_var = lambda _,N: olli(N, r, K, l)
  label = "l = "+str(l)
  olli_sol = slv.solve_ivp(olli_var, t_range, [N0], method='BDF', 
                         atol=atol, rtol=rtol, first_step=0.1)
  plt.plot(olli_sol.t, olli_sol.y[0], label=label)
  plt.text(1.5, l+0.5, label)
  plt.hlines(l, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")

plt.ylim(0, 21)
plt.hlines(K, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.text(1, 20.2, "N = K")
plt.legend()
None
```

C l=2 график выглядит как-то очень странно.

## 3 Построение неавтономной модели

Мой вариант - популяционная модель Олли. 

$$ \frac{\partial N}{\partial t} = rN \left (1 - \frac{N}{K} \right )(N - l) $$

где 0 < *l* < *K*, *r* > 0.

Надо добавить к модели ограниченный член |F(N, t)| < a = const. Добавим F(t) = a*sin(t).

$$ \frac{\partial N}{\partial t} = rN \left (1 - \frac{N}{K} \right )(N - l) + a*sin(t) $$

По идее, если a = 1, это не сильно изменит уже полученные решения. Но если a сравнимо с $l$, то есть $a = l - \varepsilon $, поведение функции заметно изменится. Посмотрим это на графике.


```python
olli_non_autonomous  = lambda N,t,r,K,l,a: r*N*(1 - N/K)*(l - N) - a*np.sin(t)

plt.figure(figsize=(10, 7))
plt.xlabel('t')
plt.ylabel('N')

r = 1
K = 20
l = 10
a = 1

N0 = 15

func_non_autonomous = lambda t, N: olli_non_autonomous(N, t, r, K, l, a)
t_range = [0, 4.4]

for N0 in [5, 8, 12, 15, 18, 22]:
#  olli_var = lambda _,N: olli(N, r, K, l)
  label = "N0 = "+str(N0)
  olli_sol = slv.solve_ivp(func_non_autonomous, t_range, [N0], method='BDF', 
                         atol=1, rtol=1, first_step=0.05)
  plt.plot(olli_sol.t, olli_sol.y[0], label=label)
  #plt.text(1.5, l+0.5, label)
  #plt.hlines(l, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")

plt.ylim(0, 21)

plt.hlines(K, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.text(1, 20.2, "N = K")

plt.hlines(l, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.text(1, 20.2, "N = l")

plt.legend()
None
```

В целом, в промежутках при $0 > N0 > l$ и $l > N0 > K$ уравнение сходится, как и при автономном уравнении. При $N0=18$, т.е. при $N0$, которое ближе к $K$, чем к $l$, уравнение начинает расходится. То же самое при $N0 > K$, т.е. при $N0 = 22$, но это не отличается от стационарного случая. 

Посмотрим, что произойдет при члене $F(t, N)$, который сильнее колеблется, т.е. при $F(t) = a*sin(t), a > 1$

```python
olli_non_autonomous  = lambda N,t,r,K,l,a: r*N*(1 - N/K)*(l - N) - a*np.sin(t)

plt.figure(figsize=(12, 7))
plt.xlabel('t')
plt.ylabel('N')
plt.title("Olli, a = 3")

r = 1
K = 20
l = 10
a = 3

N0 = 15

t_range = [0, 30.0]

for N0 in [5, 8, 12, 15, 18, 22]:
  func_non_autonomous = lambda t, N: olli_non_autonomous(N, t, r, K, l, a)
  label = "N0 = "+str(N0) + ", a = "+str(a)
  olli_sol = slv.solve_ivp(func_non_autonomous, t_range, [N0], method='BDF', 
                          atol=1, rtol=1, first_step=0.05)
  plt.plot(olli_sol.t, olli_sol.y[0], label=label)

plt.ylim(0, 21)

plt.hlines(K, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.text(1, 20.2, "N = K")

plt.hlines(l, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.text(1, 20.2, "N = l")

plt.legend()
None
```

```python
olli_non_autonomous  = lambda N,t,r,K,l,a: r*N*(1 - N/K)*(l - N) - a*np.sin(t)

plt.figure(figsize=(12, 7))
plt.xlabel('t')
plt.ylabel('N')
plt.title("Olli, a = 10")

r = 1
K = 20
l = 10
a = 10

N0 = 15

t_range = [0, 30.0]

for N0 in [5, 8, 12, 15, 18, 22]:
  func_non_autonomous = lambda t, N: olli_non_autonomous(N, t, r, K, l, a)
  label = "N0 = "+str(N0) + ", a = "+str(a)
  olli_sol = slv.solve_ivp(func_non_autonomous, t_range, [N0], method='BDF', 
                          atol=1, rtol=1, first_step=0.01)
  plt.plot(olli_sol.t, olli_sol.y[0], label=label)

plt.ylim(0, 21)

plt.hlines(K, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.text(1, 20.2, "N = K")

plt.hlines(l, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.text(1, 20.2, "N = l")

plt.legend()
None
```

При a = 10 все решения стали расходиться. Кстати, проверим, что при a = 1 решения не расходятся, просто на большем промежутке. 

```python
plt.figure(figsize=(30, 7))
plt.xlabel('t')
plt.ylabel('N')

r = 1
K = 20
l = 10
a = 1

N0 = 15

func_non_autonomous = lambda t, N: olli_non_autonomous(N, t, r, K, l, a)
t_range = [0, 500]

for N0 in [5, 8, 12, 15, 18, 22]:
#  olli_var = lambda _,N: olli(N, r, K, l)
  label = "N0 = "+str(N0)
  olli_sol = slv.solve_ivp(func_non_autonomous, t_range, [N0], method='BDF', 
                         atol=1, rtol=1, first_step=0.1)
  plt.plot(olli_sol.t, olli_sol.y[0], label=label)
  #plt.text(1.5, l+0.5, label)
  #plt.hlines(l, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")

plt.ylim(0, 17)

plt.hlines(l, t_range[0], t_range[1], color="grey", linewidth=1, linestyle="--")
plt.text(1, 20.2, "N = l")

plt.legend()
None
```
